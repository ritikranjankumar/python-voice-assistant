```python
import speech_recognition as sr
import pyttsx3
import datetime
import re

# Initialize the recognizer and the text-to-speech engine
recognizer = sr.Recognizer()
engine = pyttsx3.init()

# Function to listen to the microphone and recognize speech
def listen():
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
    try:
        print("Recognizing...")
        text = recognizer.recognize_google(audio)
        print(f"User said: {text}")
        return text
    except sr.UnknownValueError:
        print("Sorry, I did not understand that.")
        return ""
    except sr.RequestError:
        print("Sorry, there is an issue with the recognition service.")
        return ""

# Function to respond with text-to-speech
def respond(text):
    engine.say(text)
    engine.runAndWait()

# Function to provide answers based on user input
def get_response(command):
    command = command.lower()
    
    if 'time' in command:
        now = datetime.datetime.now()
        return f"The current time is {now.strftime('%H:%M')}."
    
    elif re.search(r'(\d+)\s*(\+|\-|\*|/)\s*(\d+)', command):
        match = re.search(r'(\d+)\s*(\+|\-|\*|/)\s*(\d+)', command)
        num1, operator, num2 = int(match.group(1)), match.group(2), int(match.group(3))
        
        if operator == '+':
            return f"The result is {num1 + num2}."
        elif operator == '-':
            return f"The result is {num1 - num2}."
        elif operator == '*':
            return f"The result is {num1 * num2}."
        elif operator == '/':
            return f"The result is {num1 / num2}."
    
    elif 'hello' in command or 'hi' in command:
        return "Hello! How can I assist you today?"
    
    elif 'exit' in command:
        return "Goodbye!"
    
    else:
        return "I'm not sure how to respond to that."

# Main function to run the voice assistant
def voice_assistant():
    while True:
        command = listen()
        if command:
            response = get_response(command)
            print(response)
            respond(response)
            if 'goodbye' in response.lower():
                break

# Run the voice assistant
voice_assistant()
```
